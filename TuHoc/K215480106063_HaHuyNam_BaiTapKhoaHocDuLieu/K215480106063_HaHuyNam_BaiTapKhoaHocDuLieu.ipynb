```python
#Bài 0: vẽ biểu đồ quãng đường
# Import thư viện matplotlib
import matplotlib.pyplot as plt

# Dữ liệu thời gian (x) và quãng đường (y)
time = [0, 1, 2, 3, 4, 5]
distance = [0, 10, 20, 35, 55, 80]

# Vẽ biểu đồ
plt.plot(time, distance, marker='o', linestyle='-', color='r')

# Thêm tiêu đề và nhãn trục
plt.title('Biểu đồ quãng đường theo thời gian')
plt.xlabel('Thời gian (giây)')
plt.ylabel('Quãng đường (mét)')

# Hiển thị biểu đồ
plt.grid()
plt.show()
```

```python
#Bài 1: Vẽ biểu đồ hiển thị y=x^2 với x từ 0 đên 10 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)  
y = x**2                     

plt.plot(x, y)
plt.title("Đồ thị y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#Bài 3: Vẽ y=2x^3 và y=(3x-1)^2 trên 1 biểu đồ nhưng màu khác để phân biệt 
y2 = 2*x + 3               
y3 = (3*x - 1)**2          

plt.plot(x, y2, color="blue", label="y = 2x+3")
plt.plot(x, y3, color="red", label="y = (3x-1)^2")
plt.title("Biểu đồ các đường")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()                
plt.show()
```

```python
#Bài 5: Biểu đồ phân tán 50 điểm ngẫu nhiên trục (x,y)
sizes = y                   # Sử dụng giá trị y làm kích thước điểm
colors = x                  # Sử dụng giá trị x làm màu sắc

plt.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.7)
plt.title("Biểu đồ phân tán y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="Giá trị x")  # Thêm thanh màu
plt.show()
```

```python
#Bài 7: Hiển thị điểm 5 môn học
mon_hoc = ["Toán", "Vật lý", "Hóa học", "Ngữ văn", "Tiếng Anh"]
diem = [85, 90, 78, 88, 92]  # Ví dụ điểm số

plt.bar(mon_hoc, diem, color="orange")
plt.title("Điểm số theo môn học")
plt.ylabel("Điểm số")
plt.show()
```

```python
#Bài 8: Vẽ biểu đồ 1000 điểm 
data = np.random.randn(50)  # Tạo dữ liệu ngẫu nhiên theo phân phối chuẩn

plt.hist(data, bins=30, color="skyblue", edgecolor="black")
plt.title("Biểu đồ tần số của dữ liệu ngẫu nhiên")
plt.xlabel("Giá trị")
plt.ylabel("Tần số")
plt.show()
```
