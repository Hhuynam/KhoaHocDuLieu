#Phần 4.Lấy mẫu lại dữ liệu: Chuyển từ dữ liệu theo giờ thành trung bình nhiệt độ theo ngày

```python
file_path = r'D:\HocKy3_2024_2025\KhoaHocDuLieu\PhanTichDuLieuThoiTiet\archive\temperature.csv'
data = pd.read_csv(file_path)
print(data.head())
```

```python
data['datetime'] = pd.to_datetime(data['datetime'])
print(data.dtypes)
```

```python
data.set_index('datetime', inplace=True)
daily_average = data.resample('D').mean()
print(daily_average.head())
```

#5.Nội suy dữ liệu bị thiếu để điền nhiệt độ bị khuyết

```python
print(data.isnull().sum())
missing_per_column = data.isnull().sum()
print(missing_per_column)
total_missing = data.isnull().sum().sum()
print(f"Tổng số giá trị bị thiếu: {total_missing}")
```

```python
data_xulythieu = data.interpolate(method='linear')
print(data_xulythieu.isnull().sum())
print(data_xulythieu.head())
```

#6.Tính toán xu hướng thời gian: Tính nhiệt độ trung bình theo tháng và vẽ biểu đồ

```python
%pip install matplotlib
import matplotlib 
import matplotlib.pyplot as plt
print("Phiên bản matplotlib:", matplotlib.__version__)
```

```python
daily_average_temp = data.resample('D').mean()
print(daily_average_temp.head())
```

```python
vancouver_data = daily_average_temp['Vancouver']
los_angeles_data = daily_average_temp['Los Angeles']
print(vancouver_data.head())
print(los_angeles_data.head())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(vancouver_data.index, vancouver_data, label='Vancouver', color='blue')
plt.plot(los_angeles_data.index, los_angeles_data, label='Los Angeles', color='red')
plt.title('Xu hướng nhiệt độ trung bình theo ngày - Vancouver vs Los Angeles', fontsize=16)
plt.xlabel('Ngày', fontsize=12)
plt.ylabel('Nhiệt độ trung bình (°C)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
